```python
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
#from sklearn. metrics import
df=load_digits()
x=df['data']
y=df['target']
model=StandardScaler()
x_scaled=model.fit_transform(x)
pca=PCA(0.95)
x_pca=pca.fit_transform(x_scaled)
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2)
model1=LogisticRegression()
model2=SVC()
model3=KNeighborsClassifier()
model4=DecisionTreeClassifier(max_depth=7)
model5=RandomForestClassifier()
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)
model5.fit(x_train,y_train)
print("LogisticRegression Accuracy = ",model1.score(x_test,y_test))
print("SVM Accuracy = ",model2.score(x_test,y_test))
print("KNN Accuracy = ",model3.score(x_test,y_test))
print("Decision Tree Accuracy = ",model4.score(x_test,y_test))
print("Random Forest Accuracy = ",model5.score(x_test,y_test))
print("The best Accuracy of all above is SVM i.e = ",model2.score(x_test,y_test))
plt.figure(figsize=(250,90))
plot_tree(model4,filled=True,fontsize=8)
plt.title('Decision tree')
y1_pred=model1.predict(x_test)
y2_pred=model2.predict(x_test)
y3_pred=model3.predict(x_test)
y4_pred=model4.predict(x_test)
y5_pred=model5.predict(x_test)
cm1=confusion_matrix(y_test,y1_pred)
cm2=confusion_matrix(y_test,y2_pred)
cm3=confusion_matrix(y_test,y3_pred)
cm4=confusion_matrix(y_test,y4_pred)
cm5=confusion_matrix(y_test,y5_pred)
print("Confusion matrix for logistic regression")
print(cm1)
print("Confusion matrix for SVM")
print(cm2)
print("Confusion matrix for KNN")
print(cm3)
print("Confusion matrix for Decision tree")
print(cm4)
print("Confusion matrix for Random forest")
print(cm5)
plt.figure(figsize=(10,7))
sns.heatmap(cm1,annot=True)
plt.title('confusion matrix for logistic regression')
plt.xlabel('Predicted-1')
plt.ylabel('Truth-1')
plt.show()
sns.heatmap(cm2,annot=True)
plt.title('confusion matrix for SVM')
plt.xlabel('Predicted-2')
plt.ylabel('Truth-2')
plt.show()
sns.heatmap(cm3,annot=True)
plt.title('confusion matrix for KNN')
plt.xlabel('Predicted-3')
plt.ylabel('Truth-3')
plt.show()
sns.heatmap(cm4,annot=True)
plt.title('confusion matrix for Decision tree')
plt.xlabel('Predicted-4')
plt.ylabel('Truth-4')
plt.show()
sns.heatmap(cm5,annot=True)
plt.title('confusion matrix for Random forest')
plt.xlabel('Predicted-5')
plt.ylabel('Truth-5')
plt.show()
```
